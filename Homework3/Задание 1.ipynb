```python
import pandas
import pandas as pd 


df = pd.read_csv("Electronic_sales_Sep2023-Sep2024.csv", sep=",")
payment_counts = df.groupby(['Customer ID', 'Payment Method']).size().reset_index(name='count')

preferred_payment = payment_counts.loc[payment_counts.groupby('Customer ID')['count'].idxmax()]

prefpayment = preferred_payment[['Customer ID', 'Payment Method']]

prefpayment
```

```python
total_spending = df.groupby('Customer ID')['Total Price'].sum().reset_index()
total_spending.head()
```

```python
accessory_spending = df.groupby('Customer ID')['Add-on Total'].sum().reset_index()
accessory_spending
```

```python

final_customer_analysis = prefpayment.merge(total_spending, on='Customer ID').merge(accessory_spending, on='Customer ID')
final_customer_analysis.head(10)
```
